```python
import pandas as pd
import numpy as np
import bokeh
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from bokeh.layouts import row
from bokeh.io import show, output_file
from bokeh.models.tools import HoverTool 
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
pd.set_option('display.float_format',lambda x : '%.2f' % x)
output_notebook()
```

```python
df = pd.read_csv('data.csv',engine='python')
df.head()
```

```python
Date = pd.to_datetime(pd.Series(['2019-05','2019-06','2019-07','2019-08','2019-09','2019-10','2019-11','2019-12',
       '2020-01','2020-02','2020-03','2020-04','2020-05']), format = '%Y-%m')
```

```python
month = ['M1','M2','M3','M4','M5','M6','M7','M8','M9','M10','M11','M12','M13']
## 按月分資料 key=month ,value=data
data={}
for i in range(1,14): 
    data[month[i-1]]=df[['AMT_EXCHANGE_A_M%s'%i,'AMT_EXCHANGE_B_M%s'%i,'AMT_EXCHANGE_C_M%s'%i,
                        'CNT_EXCHANGE_A_M%s'%i,'CNT_EXCHANGE_B_M%s'%i,'CNT_EXCHANGE_C_M%s'%i,'AMT_INVST_M%s'%i,
                        'CNT_INVST_M%s'%i,'AMT_LOAN_D_M%s'%i,'AMT_LOAN_E_M%s'%i,'AMT_LOAN_F_M%s'%i,'FLG_WEB_M%s'%i,
                        'YN_WEB_ACTIVE_M%s'%i,'YN_SEC_ACC_M%s'%i,'CNT_WEB_LOGIN_M%s'%i,'CNT_INTERACT_G_M%s'%i,
                        'CNT_INTERACT_H_M%s'%i,'AMT_INTERACT_G_M%s'%i,'AMT_INTERACT_H_M%s'%i,'AUM_I_M%s'%i,
                        'AUM_J_M%s'%i,'AUM_K_M%s'%i,'AUM_L_M%s'%i,'AUM_M_M%s'%i,'AUM_M%s'%i]]
```

```python
data[month[0]] #M1
```

```python
#同一月份各類別資料
catg = ['AMT_EXCHANGE_A','AMT_EXCHANGE_B','AMT_EXCHANGE_C' ,'CNT_EXCHANGE_A'     
,'CNT_EXCHANGE_B' ,'CNT_EXCHANGE_C','AMT_INVST','CNT_INVST' ,'AMT_LOAN_D','AMT_LOAN_E'          
,'AMT_LOAN_F' ,'FLG_WEB' ,'YN_WEB_ACTIVE','YN_SEC_ACC' ,'CNT_WEB_LOGIN' ,'CNT_INTERACT_G'      
,'CNT_INTERACT_H','AMT_INTERACT_G' ,'AMT_INTERACT_H','AUM_I','AUM_J' ,'AUM_K','AUM_L' ,'AUM_M' ,'AUM']
cat_data={}
for i in range(len(catg)):
    j=[]
    for k in range(1,14):
        m = [catg[i-1]+'_M%s'%k]
        j.extend(m)
    cat_data[catg[i-1]] = df[j]
```

```python
cat_data[catg[0]] #AMT_EXCHANGE_A
```

```python
## AMT_EXCHANGE_A 平均
a = list(cat_data[catg[0]].mean())
p = figure( x_axis_type="datetime",plot_height=350, plot_width=800)
p.line(Date, a)
hover = HoverTool(tooltips=[('mean','@y{0.00}')])
p.add_tools(hover)
```

```python
#換匯金額通路A~C
combined = []
color = ['red','green','blue']
p = figure( x_axis_type="datetime",plot_height=350, plot_width=800)
for i in range(3):
    combined.append(list(cat_data[catg[i]].mean()))
    p.line(Date, combined[i],color=color[i] , legend_label=catg[i],muted_color=color[i], muted_alpha=0.2)
    hover = HoverTool(tooltips=[('mean','@y{0.00}')])
p.add_tools(hover)
p.legend.location = "top_left"
p.legend.click_policy = "mute"
show(p)
```

```python
output=[]
for j in range(3):
    for k in range(len(combined[0])):
        output.append([str(Date[k])[:7], combined[j][k],catg[j]])
output=pd.DataFrame(output,columns=['Date','Mean','Category'])
output.to_csv('output.csv', index=False)
```
